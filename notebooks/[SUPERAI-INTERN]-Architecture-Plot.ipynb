```python
from graphviz import Digraph

dot = Digraph()

# Add nodes
dot.node('I', 'Input\n(224x224x3)')
dot.node('C1', 'Conv\n(3x3)')
dot.node('R1', 'ReLU')
dot.node('P1', 'MaxPool')
dot.node('FC', 'Fully Connected')

# Add edges correctly
dot.edges([('I', 'C1'), ('C1', 'R1'), ('R1', 'P1'), ('P1', 'FC')])

# Render
dot.render('arch', format='png', cleanup=True)  # cleanup removes .gv file
```
